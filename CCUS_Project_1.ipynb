```python
def stat_summary(df, cate_col, num):
    """
    Compute mean, median, and standard deviation for numerical columns grouped by each categorical column.

    Parameters:
        df (pd.DataFrame): The dataset.
        cate_cols (list): List of categorical column names.
        num_cols (list): List of numerical column names.

    Returns:
        dict: A dictionary with categorical columns as keys and group statistics DataFrames as values.
    """
    group_stats = {}

    for cat_col in cate_col:
        print(f"\nStatistics grouped by: {cat_col}")
        grouped = df.groupby(cat_col)[nums].agg(['mean', 'median', 'std'])
        grouped.columns = ['_'.join(col).strip() for col in grouped.columns]  # Flatten column names
        group_stats[cat_col] = grouped.reset_index()
        print(group_stats[cat_col])

    return group_stats

group_statistics = stat_summary(df, cate_col, nums)
```

**Categorical - categorical Relationship**

```python
#creating crosstab
contingency_table = pd.crosstab(df['region'],df['rocktype'])
print('Crosstab')
print(contingency_table)
```

```python
# performing chi square test
from scipy.stats import chi2_contingency
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("chi sqaure test result")
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 8))  # Create one figure with 2 subplots (side by side)

# Observed Frequencies
sns.heatmap(contingency_table, annot=True, fmt=".0f", cmap="coolwarm", ax=axes[0])
axes[0].set_title("Observed Frequencies")

# Expected Frequencies
sns.heatmap(expected, annot=True, fmt=".1f", cmap="coolwarm", annot_kws={"size": 10}, ax=axes[1])
axes[1].set_title("Expected Frequencies")

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

Note-
1. High value of Chi ssqaure test suggest that, the two features columns and rocktype are not independent, and it does make sense, as rocktype varies with the regions.

2.  The P- value is much less than the confidence level of 5, which rejects the null hypothesis and supports our 1st conlusion of independence of two features.

3. Degree of Freedom is high, 72, which is usuualy high data sets with large numbers of columns.

4. The expected value suggests what should be the vlaues if the null hypothesis is validated.

```python
from scipy.stats import skew

# Select numerical columns (if not already identified)
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Check skewness and plot histograms
for col in numerical_cols:
    skewness = skew(df[col].dropna())
    print(f"{col}: Skewness = {skewness}")
    plt.hist(df[col].dropna(), bins=30)
    plt.title(f"Distribution of {col}")
    plt.show()


```

```python
# Log transformation function
def log_transform(df, columns):
    """
    Apply log transformation to specified columns in a DataFrame.

    Parameters:
        df (pd.DataFrame): The dataset.
        columns (list): List of column names to transform.

    Returns:
        pd.DataFrame: DataFrame with log-transformed columns added.
    """
    df_transformed = df.copy()
    for col in columns:
        if (df[col] > 0).all():  # Use np.log for strictly positive values
            df_transformed[col + '_log'] = np.log(df[col])
        elif (df[col] >= 0).all():  # Use np.log1p for non-negative values
            df_transformed[col + '_log'] = np.log1p(df[col])
        else:
            print(f"Skipping {col} due to negative values.")
    return df_transformed

# Identify skewed columns for transformation
skewed_cols = ['permeability_md', 'pressure_psig', 'co2_density']  # Example columns; adjust based on analysis
df_transformed = log_transform(df, skewed_cols)
```

```python
# Check skewness of transformed features
for col in skewed_cols:
    log_col = col + '_log'
    if log_col in df_transformed.columns:
        skewness = skew(df_transformed[log_col].dropna())
        print(f"{log_col}: Skewness after transformation = {skewness}")
        plt.hist(df_transformed[log_col].dropna(), bins=30)
        plt.title(f"Distribution of {log_col}")
        plt.show()
```

**Scaling and normalizing the data set**

```python
from sklearn.preprocessing import StandardScaler

def standardize_data(df, columns):
    """
    Standardizes the specified numerical columns in the DataFrame.

    Parameters:
        df (pd.DataFrame): The dataset.
        columns (list): List of numerical columns to scale.

    Returns:
        pd.DataFrame: DataFrame with standardized features added.
    """
    scaler = StandardScaler()
    df_scaled = df.copy()

    # Apply standardization
    scaled_features = scaler.fit_transform(df[columns])
    scaled_df = pd.DataFrame(scaled_features, columns=[col + '_scaled' for col in columns], index=df.index)

    # Concatenate scaled features with the original DataFrame
    df_scaled = pd.concat([df_scaled, scaled_df], axis=1)
    return df_scaled
```

```python
# Inspect relevant columns
columns_to_check = ['pressure_psig', 'frac_pressure', 'stored_mid', 'porosity', 'permeability_md', 'brine_salinity', 'thickness_m']
print(df[columns_to_check].isnull().sum())  # Check for missing values
print(df[columns_to_check].describe())  # Check for zeros or unusual values
```

```python
df['frac_pressure'] = df['frac_pressure'].replace(0, 1e-6)
df['stored_mid'] = df['stored_mid'].replace(0, 1e-6)
df['thickness_m'] = df['thickness_m'].replace(0, 1e-6)
df['brine_salinity'] = df['brine_salinity'].replace(0, 1e-6)
```

```python
# Feature interaction

def add_interaction_features(df):
    """
    Adds interaction features to the dataset.
    """
    df['pressure_to_depth'] = df['pressure_psig'] / df['depth']
    df['porosity_to_permeability'] = df['porosity'] / (df['permeability_md'] + 1e-6)  # Avoid division by zero
    df['normalized_efficiency'] = df['stored_mid'] / df['thickness_m']
    return df

# Apply the updated function
df_interactions = add_interaction_features(df)

# Verify interaction features
df_interactions[['pressure_to_depth', 'porosity_to_permeability', 'normalized_efficiency']].head()
```

```python
# Debug Leakage_Risk
df['Leakage_Risk'] = df['pressure_psig'] / (df['frac_pressure'] * df['stored_mid'])
print(df['Leakage_Risk'].isnull().sum())  # Check if NaNs persist

# Debug Reservoir_Suitability
df['Reservoir_Suitability'] = df['porosity'] * df['permeability_md'] / df['brine_salinity']
print(df['Reservoir_Suitability'].isnull().sum())  # Check if NaNs persist

# Debug Pressure_Gradient
df['Pressure_Gradient'] = df['pressure_psig'] / df['thickness_m']
print(df['Pressure_Gradient'].isnull().sum())  # Check if NaNs persist
```

```python
def add_domain_features(df):
    """
    Adds domain-specific features for CO2 storage site analysis with safeguards.
    """
    # Replace zero or NaN values to avoid division errors
    df['frac_pressure'] = df['frac_pressure'].replace(0, 1e-6).fillna(1e-6)
    df['stored_mid'] = df['stored_mid'].replace(0, 1e-6).fillna(1e-6)
    df['thickness_m'] = df['thickness_m'].replace(0, 1e-6).fillna(1e-6)
    df['brine_salinity'] = df['brine_salinity'].replace(0, 1e-6).fillna(1e-6)

    # Add features
    df['Leakage_Risk'] = df['pressure_psig'] / (df['frac_pressure'] * df['stored_mid'])
    df['Reservoir_Suitability'] = df['porosity'] * df['permeability_md'] / df['brine_salinity']
    df['Pressure_Gradient'] = df['pressure_psig'] / df['thickness_m']
    return df

# Apply the function
df = add_domain_features(df)
```

# Model Training

Note- Creating another column named Suitable_site, to asses the sites whether they are suitable for storing the CO2 or not,
So, if the reservoir suitability is above a thresold value, we can be positive about storing at that particular site.

```python
# Convert categorical columns to 'category' dtype
categorical_columns = ['formation', 'containment_summary']
for col in categorical_columns:
    df[col] = df[col].astype('category')
```

```python
# Step 1: Define threshold for stability
threshold = 0.5  # Example threshold; adjust based on domain knowledge.

# Step 2: Create binary column for stability
df['Suitable_Site'] = (df['Reservoir_Suitability'] > threshold).astype(int)

# Step 3: Define X and y
y = df['Suitable_Site']
x = df.drop(columns=['Suitable_Site', 'Leakage_Risk', 'Reservoir_Suitability'])
```

```python
# Split the dataset
from sklearn.model_selection import train_test_split

X = df.drop(columns=['Suitable_Site'])  # Replace 'Suitable_Site' with your actual target column
y = df['Suitable_Site']  # Target column

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print("Features Data Types:\n", X.dtypes)
print("Target Data Type:", y.dtype)
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# Replace inf and -inf with NaN
X_train.replace([np.inf, -np.inf], np.nan, inplace=True)
X_test.replace([np.inf, -np.inf], np.nan, inplace=True)

# Handle numerical columns
numerical_cols = X_train.select_dtypes(include=["float64", "int64"]).columns
X_train[numerical_cols] = X_train[numerical_cols].fillna(X_train[numerical_cols].mean())
X_test[numerical_cols] = X_test[numerical_cols].fillna(X_test[numerical_cols].mean())

# Handle categorical columns
categorical_cols = X_train.select_dtypes(include=["category", "object"]).columns
X_train[categorical_cols] = X_train[categorical_cols].fillna(X_train[categorical_cols].mode().iloc[0])
X_test[categorical_cols] = X_test[categorical_cols].fillna(X_test[categorical_cols].mode().iloc[0])
```

```python
print("Missing values in X_train:", X_train.isnull().sum().sum())
print("Missing values in X_test:", X_test.isnull().sum().sum())
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
import numpy as np

# Initialize StratifiedKFold
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Store cross-validation scores
cv_scores = []

# Perform cross-validation
for train_index, val_index in kf.split(X_train, y_train):
    X_cv_train, X_cv_val = X_train.iloc[train_index], X_train.iloc[val_index]
    y_cv_train, y_cv_val = y_train.iloc[train_index], y_train.iloc[val_index]

    # Train the model
    xgb_model.fit(X_cv_train, y_cv_train)

    # Predict and evaluate
    y_cv_pred = xgb_model.predict(X_cv_val)
    cv_scores.append(accuracy_score(y_cv_val, y_cv_pred))

print("Cross-Validation Accuracy:", np.mean(cv_scores))
```
